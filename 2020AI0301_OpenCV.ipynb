## openCVで画像を操作する

```
# 画像をネットからダウンロード
import urllib.request as req
url = "https://raw.githubusercontent.com/masatokg/book-mlearn-gyomu/master/src/ch3/cv2io/test.jpg"
# ファイルを保存
savefile = "test.png"
req.urlretrieve(url, savefile)

# 保存したファイルをopenCVで読み込む
import cv2
img = cv2.imread(savefile)
# 画像データ（ピクセル情報）を表示
print(img)
```

### わざと間違えてimreadの戻り値をみる

```
# 存在しないファイルパスを指定して読込命令をすると、戻り値はNoneになる
ng_img = cv2.imread("存在しないファイル")
print(ng_img)
```

### 読み込んだ画像データ（ピクセル情報）をインライン表示（Notebook上に表示）

```
# matplotlib にインライン表示宣言をする
%matplotlib inline
import matplotlib.pyplot as plt
# 画像データのプロットを作る(ピクセル情報そのまま)
plt.imshow(img)
# プロットを表示
plt.show()
# ピクセル情報を変換してプロットを作る（ BGR -> RGB で指定）
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### 画像を保存する

```
# 保存するファイル名
out_png = "out.png"
# openCVの画像保存メソッド
cv2.imwrite(out_png, img)
```

### メモリ軸をオフにしたプロットを作って表示

```
# メモリ軸をオフにしたプロットを作る
plt.axis("off")
plt.imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB) )
# 作ったプロットを表示
plt.show()
```

## 画像をリサイズ（拡大・縮小）する

```
# 読み込んだ画像 img をリサイズ
im2 = cv2.resize( img, ( 600,300 ) ) # 横幅w 600, 縦幅h 300にリサイズして代入
# リサイズした画像をファイルに保存
cv2.imwrite("out-resize.png", im2)
# 画像のプロットを作って表示
plt.imshow( cv2.cvtColor( im2, cv2.COLOR_BGR2RGB ) )
plt.show() # 表示
```

### 画像の切り取り（抽出）を行う
### 切り取りたい領域は、y座標, x座標を 2次元配列をイメージした要領で指定する
### [開始行:終端行, 開始列:終端列]

```
# 画像を切り取り（抽出して新たな画像・変数に代入）
im3 = img[150:450, 150:450]
cv2.imwrite( "cut.png", im3 )
plt.imshow( cv2.cvtColor( im3, cv2.COLOR_BGR2RGB ) )
plt.show()

plt.imshow( cv2.cvtColor( img, cv2.COLOR_BGR2RGB ) )
plt.show()
```

### 切り取った画像をリサイズ

```
im4 = cv2.resize( im3, (400, 400) )
cv2.imwrite( "cut-resize.png", im4)
plt.imshow( cv2.cvtColor( im4, cv2.COLOR_BGR2RGB ) )
plt.show()
```
